```python
import pandas as pd
```

```python
df = pd.read_csv('./raw_data_joo.csv', encoding='euc-kr')
```

```python
df.tail()
```

```python
cnykrw = pd.read_csv('./데이터/CNY_KRW 일별 2010~.csv')[['날짜','종가']].sort_values('날짜')
```

```python
cnykrw['날짜'] = pd.to_datetime(cnykrw['날짜']).dt.strftime('%Y-%m-%d')
```

```python
cnykrw=cnykrw.rename(columns = {'종가':'위안화원'})
```

```python
cnykrw
```

```python
jptkrw = pd.read_csv('./데이터/JPY_KRW 과거 데이터.csv')[['날짜','종가']].sort_values('날짜')
```

```python
jptkrw['날짜'] = pd.to_datetime(jptkrw['날짜']).dt.strftime('%Y-%m-%d')
```

```python
jptkrw=jptkrw.rename(columns = {'종가':'엔원'})
```

```python
jptkrw.isna().sum()
```

```python
df1 = pd.merge(df,cnykrw, how='left',on='날짜')
```

```python
df2 = pd.merge(df1,jptkrw, how='left',on='날짜')
```

```python
df2 = df2.bfill()
```

```python
df2
```
