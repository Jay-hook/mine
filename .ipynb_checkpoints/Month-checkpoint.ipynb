```python
m = M.drop(['날짜','원달러종가'], axis=1)
```

```python
m.columns
```

```python
m[m.columns].corr()
```

```python
# 변수에 따라서 다중공정성없애기 
# 다중공정성으로 CD 91물, 달러엔 제외 
corr_data = m[m.columns].corr()
mask = np.triu(np.ones_like(corr_data, dtype=bool))
sns.heatmap(corr_data, annot=True, cmap='coolwarm', linewidths=.5, fmt='.2f', mask=mask)
```
